```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cancer_classification.csv')
```

```python
df.isnull().sum()
```

```python
df.describe().transpose()
```

```python
df.head()
```

```python
sns.countplot(x = 'benign_0__mal_1', data = df)
```

```python
df.corr()['benign_0__mal_1'].sort_values()
```

```python
sns.heatmap(df.corr())
```

```python
X = df.drop(columns = 'benign_0__mal_1', axis = 1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101, shuffle = True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))

model.add(Dense(15, activation = 'relu'))

# BINARY CLASSIFICATION - SIGMOID ACTIVATION
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'Adam', loss = 'binary_crossentropy' )
```

```python
model.fit(x = X_train,y = y_train, validation_data = (X_test,y_test),epochs = 600)
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss
```

```python
loss.plot() 
```

```python
predictions = model.predict(X_test)
predicted_classes = (predictions > 0.5).astype("int32")  # For binary classification
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicted_classes))
```

```python
print(confusion_matrix(y_test,predicted_classes))
```

```python
# the model misclassified 14 
```

```python
# use early stopping to get good epoch number
# min for loss, max for accuracy
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
stop_early = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 25)
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))

model.add(Dense(15, activation = 'relu'))

# BINARY CLASSIFICATION - SIGMOID ACTIVATION
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'Adam', loss = 'binary_crossentropy' )
```

```python
model.fit(x = X_train,y = y_train, validation_data = (X_test,y_test),epochs = 600,callbacks = [stop_early])
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
predictions = model.predict(X_test)
predicted_classes = (predictions > 0.5).astype("int32")  # For binary classification
```

```python
print(classification_report(y_test,predicted_classes))
```

```python
print(confusion_matrix(y_test,predicted_classes))
```

```python
# with dropouts
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(15, activation = 'relu'))
model.add(Dropout(0.5))

# BINARY CLASSIFICATION - SIGMOID ACTIVATION
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy' )
```

```python
stop_early = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 200)
```

```python
model.fit(x = X_train,y = y_train, validation_data = (X_test,y_test),epochs = 600,callbacks = [stop_early])
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
predictions = model.predict(X_test)
predicted_classes = (predictions > 0.5).astype("int32")  # For binary classification
```

```python
predicted_classes
```

```python
print(classification_report(y_test, predicted_classes))
```

```python
print(confusion_matrix(y_test,predicted_classes))
```
